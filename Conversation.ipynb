```python
print("Enter the file name")
f_name=input()
file = open(f_name)
```

```python
#creating dictionary using dict
count = 0
rowno_list= []
file = open(f_name)
for inputline in file:
    count+=1
    if count ==1:
        continue
    rowno_list.append(count)

total_value_pair= len(rowno_list)
list_new=[]

while total_value_pair!=0:
    list_new.append([[],[],[],[],[]])
    total_value_pair=total_value_pair-1
    
my_dict=dict(zip(rowno_list,list_new))

#print(my_dict)
##################################################################
import re

count_y =0;
count_n=0;
count_blank=0;
line_count=0;
file = open(f_name)
for inputline in file:
    line_count+=1
    if line_count==1:
        continue
    
    if "Y" in inputline:
         # Function to reverse a string , we reverse a string to split using Y.
        def reverse(string):
            string = "".join(reversed(string))
            return string
        
        inputline = reverse(inputline)
        inputline=inputline.split("Y",1)[1]
        inputline = reverse(inputline)  # from now input line is line before Y.
        
        accession_number = inputline[0:20]
        my_dict[line_count][0].append(accession_number) # accession number

        inputline_new = inputline[20:]
        temp = re.findall('\d*\.?\d+', inputline_new)
        my_dict[line_count][3].append(temp[-1])        # offering amount
       
        inputline=inputline.replace(" ","p")           # extarcting compenasation % and interst %
        inputline=inputline.replace("ppercent","%")
        res = re.findall('\d*\.?\d*%', inputline)
        
        
        res_new=[]
        for x in res:
            if x !="%":
                res_new.append(x)
                
        #print(res_new) # resnew is a list contains compenasation % and interst %
       
        if len(res_new)==1:                 # populating dict
            c= res_new[0].replace("%","")
            my_dict[line_count][1].append(c)
            my_dict[line_count][2].append("NONE")
            my_dict[line_count][4].append((float(temp[-1])/100)*float(c))
        
        elif len(res_new)!=0 and len(res_new)!=1:
            a=res_new[0]
            b=res_new[-1]
            a=a.replace("%","")
            b=b.replace("%","")
            my_dict[line_count][1].append(a)
            my_dict[line_count][2].append(b)  
            my_dict[line_count][4].append((float(temp[-1])/100)*float(a))

    
    
    
    
    elif "N" in inputline and "Y" not in inputline :
        # Function to reverse a string , we reverse a string to split using "N"
        def reverse(string):
            string = "".join(reversed(string))
            return string
        #print(inputline)
        inputline = reverse(inputline)
        inputline=inputline.split("N",1)[1]
        inputline = reverse(inputline)
        #print(inputline)
        
        accession_number = inputline[0:20]
        my_dict[line_count][0].append(accession_number)
        
        inputline_new = inputline[20:]
        temp = re.findall('\d*\.?\d+', inputline_new)
        my_dict[line_count][3].append(temp[-1])
        
        inputline=inputline.replace(" ","p")
        inputline=inputline.replace("ppercent","%")
        res = re.findall('\d*\.?\d*%', inputline)
        

        res_new=[]
        for x in res:
            if x !="%":
                res_new.append(x)
                
        #print(res_new)
        
        if len(res_new)==1:                 # populating dict
            c= res_new[0].replace("%","")
            my_dict[line_count][1].append(c)
            my_dict[line_count][2].append("NONE")
            my_dict[line_count][4].append((float(temp[-1])/100)*float(c))

        
        elif len(res_new)!=0 and len(res_new)!=1:
            a=res_new[0]
            b=res_new[-1]
            a=a.replace("%","")
            b=b.replace("%","")
            my_dict[line_count][1].append(a)
            my_dict[line_count][2].append(b)
            my_dict[line_count][4].append((float(temp[-1])/100)*float(a))

    
    elif "N" not in inputline and "Y" not in inputline:
        accession_number = inputline[0:20]
        my_dict[line_count][0].append(accession_number)
    
```
